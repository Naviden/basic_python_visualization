The most common way to use Matplotlib is through `pyplot`, which is typically imported as `plt`:

```python
import matplotlib.pyplot as plt
```

### Basic Plot

Let's start with a simple line plot. Here’s how you can plot a simple line graph of y versus x:

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y)
plt.show()
```

This code will open a window with a line graph of the y values against the x values.

### Adding Title and Labels

You can add titles and labels to your plot to make it more informative:

```python
plt.plot(x, y)
plt.title('Simple Plot')
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.show()
```

### Multiple Lines

You can plot multiple sets of data on the same figure by calling `plt.plot()` multiple times before calling `plt.show()`:

```python
x1 = range(2, 6)
y1 = range(5, 9)

x2 = range(2, 17)
y2 = range(0, 15)
```

```python
plt.plot(x1, y1, label='First Line')
plt.plot(x2, y2, label='Second Line')
plt.legend()
plt.show()
```

### Plotting Styles

Matplotlib allows you to adjust the style of the plot. You can change the line color, type, width, and add markers:

```python
plt.plot(x, y, color="#c79413", linestyle='--', linewidth=2, marker='+', markersize=10)
plt.show()
```

### Subplots

You can create figures with multiple subplots (axes) in Matplotlib:

```python
_, ax = plt.subplots(2)  # Creates a figure and 2 subplots (vertically)
ax[0].plot(x, y)
ax[1].plot(y, x)

# Adding titles
ax[0].set_title('First Plot')
ax[1].set_title('Second Plot')

plt.tight_layout()  # Adjusts subplot params so that subplots are nicely fit in the figure
plt.show()
```

### Other Types of Plots

Matplotlib supports a wide variety of plots, such as:

- **Bar charts**: `plt.bar()`
- **Histograms**: `plt.hist()`
- **Scatter plots**: `plt.scatter()`
- **Pie charts**: `plt.pie()`
- And many more.

### Saving Figures

You can save figures to a file using `plt.savefig()`:

```python
plt.plot(x, y)
plt.savefig('plot.png')
```

This will save the current figure to a file named `plot.png`.
